```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.svm import SVC
from functions import rbf_kernel, predict, accuracy, f1_score, MySVM, quad, confusion_matrix, roc_plot
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

# Dataset 1

```python
filename = 'data_1.h5'
f = h5py.File(filename, 'r')

X = np.array(f[list(f.keys())[0]])
y = np.array(f[list(f.keys())[1]])
print('Number of samplezs in dataset 1: ', len(y))

unique, counts = np.unique(y, return_counts=True)
print('Distribution of samples among different classes: ', counts)
```

## (i) Scatter the data

```python
plt.figure()
plt.title('Dataset 1 plot')
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.savefig('dataset1_plot.png')
plt.show()
```

## (ii) Plot the dataset with decision boundaries corresponding to the kernels used

```python
clf = SVC(kernel=quad)
clf.fit(X, y)
plot_decision_regions(X, y, clf=clf)
plt.title('Dataset 1- Decision Boundary Plot')
plt.savefig('dataset1_decision_boundary_plot.png')
plt.show()
```

## (iii) Soft margin SVM with linear kernel

```python
clf = SVC(kernel='linear')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
clf.fit(X_train, y_train)
r = predict(X_test, clf._get_coef()[0], clf.intercept_)
print('Accuracy: ', accuracy(r, y_test))
print('F1 Score: ', f1_score(y_test, r))
plot_decision_regions(X, y, clf=clf)

sv_idx = clf.support_
plt.scatter(X[sv_idx, 0], X[sv_idx, 1], c='g', marker='x', s=10)
plt.title('Dataset 1- Soft Margin SVM wit Linear Kernel')
plt.savefig('dataset1_soft_margin_linear.png')
plt.show()
```

```python
sb = confusion_matrix(y_test, r)
plt.title('Dataset 1- Confusion Matrix - Soft Margin SVM wit Linear Kernel')
plt.savefig('dataset1_soft_margin_linear_cm.png')
```

```python
roc_plot(y_test, r)
plt.title('Dataset 1- ROC Curve - Soft Margin SVM wit Linear Kernel')
plt.savefig('dataset1_soft_margin_linear_roc.png')
```


## (iv) Soft margin SVM with rbf kernel

```python
gamma = -1
kX = np.hstack((X, rbf_kernel(X, gamma)))
X_train, X_test, y_train, y_test = train_test_split(kX, y, test_size=0.20)
clf = MySVM(kernel=rbf_kernel, gamma=gamma)
clf.fit(X_train, y_train)
r = predict(X_test, clf.clf._get_coef()[0], clf.clf.intercept_)
print(accuracy(r, y_test))
print(f1_score(y_test, r))
plot_decision_regions(X, y, clf=clf)

sv_idx = clf.clf.support_
plt.scatter(X[sv_idx, 0], X[sv_idx, 1], c='g', marker='x', s=10)

plt.title('Dataset 1- Soft Margin SVM wit RBF Kernel')
plt.savefig('dataset1_soft_margin_rbf.png')
plt.show()
```

```python
sb = confusion_matrix(y_test, r)
plt.title('Dataset 1- Confusion Matrix - Soft Margin SVM wit RBF Kernel')
plt.savefig('dataset1_soft_margin_rbf_cm.png')
```

```python
roc_plot(y_test, r)
plt.title('Dataset 1- ROC - Soft Margin SVM wit RBF Kernel')
plt.savefig('dataset1_soft_margin_rbf_roc.png')
```
